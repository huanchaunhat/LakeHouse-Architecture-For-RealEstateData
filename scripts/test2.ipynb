```python
from pyspark.sql import SparkSession

# 1️⃣ Khởi tạo SparkSession
spark = (
    SparkSession.builder
    .appName("ReadLocalParquet")
    .getOrCreate()
)

# 2️⃣ Đọc file Parquet (1 file hoặc 1 thư mục chứa nhiều file)
parquet_path = "./data_silver.parquet"  # <-- thay đường dẫn thật của bạn
# Nếu bạn muốn đọc cả thư mục thì đổi thành: "/home/theanh/data/"

df = spark.read.parquet(parquet_path)

# 3️⃣ Hiển thị kết quả
df.show(20)
df.printSchema()

# 4️⃣ Dừng SparkSession
spark.stop()
```
